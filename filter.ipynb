```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import PandasTools
from rdkit.Chem import Descriptors

# Load the SDF file into a Pandas DataFrame
  # Replace with your file path
mol_supplier = Chem.SDMolSupplier(sdf_file)

# Convert to DataFrame using PandasTools
mol_df = PandasTools.LoadSDF(sdf_file)

# Optionally, you can filter and view specific columns
# For example, if you want to see the first few rows
mol_df.head()
```

```python
mol_df["mlogp"] = mol_df["ROMol"].apply(Descriptors.MolLogP)
```

```python
mol_df.head()
```

```python
mol_df['MW'] = pd.to_numeric(mol_df['MW'], errors='coerce')
mol_df['Donor'] = pd.to_numeric(mol_df['Donor'], errors='coerce')
mol_df['Acceptor'] = pd.to_numeric(mol_df['Acceptor'], errors='coerce')
mol_df['RotBonds'] = pd.to_numeric(mol_df['RotBonds'], errors='coerce')
```

```python
mol_df = mol_df[(mol_df['MW'] >= 150) & (mol_df['MW'] <= 500)]
mol_df = mol_df[(mol_df['mlogp'] >= 0) & (mol_df['mlogp'] <= 5)]
mol_df =  mol_df[(mol_df['Donor'] <= 2)]
mol_df =  mol_df[(mol_df['Acceptor'] >= 1) & (mol_df['Acceptor'] <= 8)]
mol_df = mol_df[ (mol_df['RotBonds'] <= 12)]
```

```python
from rdkit import Chem
from rdkit.Chem import PandasTools

# Let's assume you already have a DataFrame `mol_df` with RDKit molecule objects and properties

# Specify the output SDF file name
output_sdf_file = 'filtered_by_tice_rule.sdf'

# Create an SDWriter to write to the SDF file
with Chem.SDWriter(output_sdf_file) as writer:
    # Loop through the DataFrame and write each molecule along with its properties
    for index, row in mol_df.iterrows():
        mol = row['ROMol']  # Get the RDKit molecule object

        # Set custom properties for the molecule (e.g., IDNUMBER, Chemical_Name, etc.)
        mol.SetProp('IDNUMBER', str(row['IDNUMBER']))
        mol.SetProp('Chemical_Name', str(row['Chemical_Name']))
        mol.SetProp('CAS', str(row['CAS']))
        mol.SetProp('PriceCoeff', str(row['PriceCoeff']))
        mol.SetProp('MW', str(row['MW']))
        mol.SetProp('Description', str(row['Description']))
        mol.SetProp('FSP3', str(row['FSP3']))
        mol.SetProp('clogP', str(row['clogP']))
        mol.SetProp('TPSA', str(row['TPSA']))
        mol.SetProp('Acceptor', str(row['Acceptor']))
        mol.SetProp('Donor', str(row['Donor']))
        mol.SetProp('RotBonds', str(row['RotBonds']))

        # Write the molecule to the SDF file
        writer.write(mol)

print(f"SDF file written to: {output_sdf_file}")
```
